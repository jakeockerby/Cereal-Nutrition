```python
# Name: Name of cereal
# mfr: Manufacturer of cereal
# A = American Home Food Products;
# G = General Mills
# K = Kelloggs
# N = Nabisco
# P = Post
# Q = Quaker Oats
# R = Ralston Purina
# type:
# cold
# hot
# calories: calories per serving
# protein: grams of protein
# fat: grams of fat
# sodium: milligrams of sodium
# fiber: grams of dietary fiber
# carbo: grams of complex carbohydrates
# sugars: grams of sugars
# potass: milligrams of potassium
# vitamins: vitamins and minerals - 0, 25, or 100, indicating the typical percentage of FDA recommended
# shelf: display shelf (1, 2, or 3, counting from the floor)
# weight: weight in ounces of one serving
# cups: number of cups in one serving
# rating: a rating of the cereals
```

```python
import pandas as pd
import numpy as np

# Loading the data
cereals = pd.read_csv('/home/jake_ockerby/Python/80_cereals/cereal.csv')
cereals
```

```python
import seaborn as sns

# Testing the relationship between rating and multiple nutritional variables
sns.regplot(x=cereals['calories'], y=cereals['rating'])
```

```python
sns.regplot(x=cereals['protein'], y=cereals['rating'])
```

```python
sns.regplot(x=cereals['fat'], y=cereals['rating'])
```

```python
sns.regplot(x=cereals['sodium'], y=cereals['rating'])
```

```python
sns.regplot(x=cereals['fiber'], y=cereals['rating'])
```

```python
sns.regplot(x=cereals['carbo'], y=cereals['rating'])
```

```python
sns.regplot(x=cereals['sugars'], y=cereals['rating'])
```

```python
# These scatterplots show that cereals with the highest rating are low in calories, low in fat,
# high in fiber and low in sugars
```

```python
import matplotlib.pyplot as plt

# Finding the highest and lowest rated manufacturers
plt.figure(figsize=(14,6))
plt.title('Mean Customer Rating per Manufacturer')
sns.barplot(x=cereals['mfr'], y=cereals['rating'])
plt.xlabel('manufacturer')
```

```python
# Testing the ratings between hot and cold cereal
plt.figure(figsize=(14,6))
plt.title('Mean Customer Rating for Hot and Cold Cereal')
sns.barplot(x=cereals['type'], y=cereals['rating'])
plt.xlabel('Temperature (H: Hot, C: Cold)')
```

```python
# From this data we can conclude that highly rated cereals are meant to be served hot,
# low in calories, high in fibre and manufactured by Nabisco
```
